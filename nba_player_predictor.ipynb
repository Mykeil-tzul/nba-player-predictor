# NBA Player Performance Predictor

## Project Goal
This project uses machine learning to predict NBA player points based on game statistics.

## Dataset
- Source: 2024/2025 NBA player game stats
- File used: `database_24_25.csv`

## Features Used
- MP (minutes played)
- FGA (field goal attempts)
- 3PA (three-point attempts)
- FTA (free throw attempts)

## Target
- PTS (points scored)

## Model
- Linear Regression

## Evaluation
- Mean Squared Error (MSE)
- Actual vs Predicted scatter plot

## 1. Import Libraries

```python
import pandas as pd
import numpy as np

print("Environment ready 🚀")
```

## 2. Load the Dataset

```python
import pandas as pd

df = pd.read_csv("database_24_25.csv")

df.head()
```

## 3. Explore the Data

```python
df.columns
```

## 4. Select Features and Target

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Features (inputs)
X = df[['MP', 'FGA', '3PA', 'FTA']]

# Target (what we predict)
y = df['PTS']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

print("Model trained 🚀")
```

## 5. Train the Model

```python
predictions = model.predict(X_test)

print(predictions[:10])
```

## 6. Make Predictions

```python
comparison = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": predictions
})

comparison.head(10)
```

## 7. Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)

print("MSE:", mse)
```

## 8. Visualize Actual vs Predicted Points

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel("Actual Points")
plt.ylabel("Predicted Points")
plt.title("Actual vs Predicted Points")

plt.show()
```

## 9. Results Summary

The linear regression model was able to learn the relationship between shot volume, minutes played, and points scored.

### Key Takeaways
- More minutes and shot attempts generally lead to higher point totals.
- The model produced a Mean Squared Error of about 12.10.
- This means the model was off by about 3.5 points on average.
- The scatter plot showed a strong positive relationship between actual and predicted points.
